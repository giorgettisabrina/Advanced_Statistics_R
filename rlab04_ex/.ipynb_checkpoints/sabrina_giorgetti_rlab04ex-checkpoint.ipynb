## Exercise 1 - Six Boxes Toy Model : inference

Write a program in R that:

1. Allows the user to insert the stone color of a randomly extracted box 
2. Prints on the standard output the probability of selecting each box
3. Plots the probability for each box as a function of the extraction step

### Insert the color of the extracted stone from a randomly extracted box

```r
E <- c()
trial <- c(0)
ext <- 100 #max number of extraction
var = readline(prompt = "Insert the color of the extracted stones (b for black, w for white, s for stop): ")
count <-0
while (var != 's' & length(E) < 100 ) {  #if another letter is inserted it will not be taken into consideration
    if (var=='b' | var=='w'){        
        E <- append(E, var)
        trial <-append(trial, tail(trial, n=1)+1)
    }
    var = readline(prompt = "Insert the stone color of the extracted stones (b for black, w for white, s for stop): ")
}

```

```r
cat("Number of extractions:", tail(trial, n=1), "\n")
cat("Vector with the number of extractions:", trial, "\n")
cat("Colors of the extractions:", E)
```

### Evaluation and print of the probability of selecting each box based on the extractions

```r
set.seed(89540) #set random seed


p_ewH <- function(j){ #probability of extracting a white stone given the H_j box
    return (j/5)
    }
p_ebH <-  function(j){ #probability of extracting a black stone given the H_j box
    return ((5-j)/5)
    }


p_row <- rep(1/6, 6)
p_row <- matrix(rep(1/6, 6), nrow=1, byrow=TRUE) 

for (i in 1:(length(E))){
    if (E[i]=='w'){
        p_row <- rbind(p_row, p_ewH(0:5)*p_row[i,]/sum(p_ewH(0:5)*p_row[i,]))

    } 
    if (E[i]=='b'){
        p_row <- rbind(p_row, p_ebH(0:5)*p_row[i,]/sum(p_ebH(0:5)*p_row[i,]))

    }
    
}

data_exp <- data.frame(trial= 0:length(E),color = c("-",E), p_row)  
names(data_exp)[3:8] <- c("H0", "H1", "H2", "H3", "H4", "H5")
options(digits=3, scipen=0)
data_exp
```

### Plots of the probability for each box as a function of the extraction step

```r
options(repr.plot.width = 25, repr.plot.height = 12)

plot_func <- function(data, p_h){
     ggplot(data, aes(x=trial,y=p_h))+ geom_point(size=3, color="red") + ylim(0,1.2) +
     geom_text(aes(label=color),  x = trial, y = 1.2, size = 5, colour = "black") +
     theme(axis.text=element_text(size=16), axis.title.x =element_text(size=15), 
           axis.title.y =element_text(size=15),plot.title=element_text(size=18,face="bold", hjust=0.5)) 
}


gg0 <- plot_func(data_exp, data_exp$H0)+  labs(title="BOX: H0", x ="Extraction", y = "Probability(box)") 
gg1 <- plot_func(data_exp, data_exp$H1)+  labs(title="BOX: H1", x ="Extraction", y = "Probability(box)")
gg2 <- plot_func(data_exp, data_exp$H2)+  labs(title="BOX: H2", x ="Extraction", y = "Probability(box)")
gg3 <- plot_func(data_exp, data_exp$H3)+  labs(title="BOX: H3", x ="Extraction", y = "Probability(box)")                                    
gg4 <- plot_func(data_exp, data_exp$H4)+  labs(title="BOX: H4", x ="Extraction", y = "Probability(box)")                                             
gg5 <- plot_func(data_exp, data_exp$H5)+  labs(title="BOX: H5", x ="Extraction", y = "Probability(box)")
                                            

grid.arrange(gg0,gg1,gg2,gg3,gg4,gg5,nrow=2)
```

An important property of the gamma distribution is the so-called reproductive property. 

Given a sequence of independent random variable $X_j∼Gamma(\alpha j,\beta)$, it follows that: 


$$ Y= \sum_{j=1}^n X_j  \rightarrow Y∼Gamma(\alpha,\beta) \;\; where \; \; \alpha= \sum_{j=1}^n \alpha_j   $$ 

if  $\alpha=m$ is an integer, a random variable from gamma distribution $Gamma(m,\beta)$ (alsoknown as Erlang distribution) can be obtained by summing m independent exponential random variables $X_j∼Exp(\beta)$:


$$Y=\beta \sum_{j=1}^n (−lnU_j)=− \beta ln \prod_{j=1}^n U_j$$

1. Write an algorithm to sample variables from an Erlang distribution Gamma(m,β)

```r
erlang <- function(n, m, beta){
    u <- matrix(runif(n*m),n,m)
        y <- apply(u, 1, function(x) -beta * log(prod(x)))
    return (y)
}

n <- 10^4
beta <- 0.5
m <- 20

samples <- erlang(n,m, beta)

#plot to see the results
options(repr.plot.width = 8, repr.plot.height = 6) 
hist(samples, freq=FALSE, ylim=c(0,0.2), xlim=c(0,25))

x<-seq(0,25,0.1)
curve(dgamma(x,m,rate=1/beta), col='red', add=TRUE)
```

The middle-square algorithm is a method to generate pseudo-random numbers, since each number is determined by its predecessor. 
Repeated enough times, the method can return as output a number with fewer digits than the previous one, and in the end it can enter a loop, generating always the same number, as showed above with the seed 87654. So in practice, depending on the random seed, it's not always a good choice.
