## Exercise 1

A publishing company has recently launched a new journal.  In order to determine how effective it is in reaching its possible audience, a market survey company selects a random sample of people from a possible target audience and interviews them.  Out of 150 interviewed people, 29 have read the last issue of the journal.

1. What kind of distribution would you assume for y, the number of people that have seen the last issue of the yournal ?
2. Assuming a uniform prior, what is the posterior distribution for y?
3.  Plot both posterior and likelihood ditributions functions

For the y we can assume a binomial distribution with n as the number of interviewed people and r the people that have read the last issue of the journal- The likelihood is then given by:

$$
P(r \mid p, n, M)=\left(\begin{array}{l}
n \\
r
\end{array}\right) p^{r}(1-p)^{n-r} \quad \text { with } \quad r \leq n
$$


Assuming a uniform prior $P(p|M)  \sim U(0,1)$ the posterior distribution for y can be obtain from Bayes's theorem: 

$$
P(p \mid n, r, M)=\frac{P(r \mid p, n, M) P(p \mid M)}{P(r \mid n, M)}
$$


$$P(p|r,n,M) = \frac{1}{Z} \cdot p^r (1-p)^{n-r}= \frac{1}{Z} \cdot P^*(p|r,n,M)  $$

with Z as the normalization factor. We can observe that the posterior is proportional to the likelihood as we have adopted a uniform prior. 

```r
n <- 150
r <-29

n.sample <- 2000
delta.p <- 1/n.sample


p <- seq(from=1/(2*n.sample),by=1/n.sample,length.out=n.sample)
prior <-dunif(p) #flat prior
likelihood <-dbinom(x=r, size=n, prob=p) #likelihood (with uniform prior it's proportional to the posterior)
p.norm <- likelihood*prior/(delta.p*sum(likelihood)) #posterior= likelihood*prior/normalization factor


#plots 
options(repr.plot.width = 15, repr.plot.height = 6) 
par(mfrow=c(1, 2))
    
#likelihood plot
plot(p, likelihood, type="l",lwd=2.5,col="darkgreen",xlim=c(0,1), ylim=c(0,1.1*max(likelihood)),
     xaxs="i",yaxs="i",xlab="p",ylab="P(r|p,n,M)")
title(main=paste("Likelihood with (n,r)=(",n, ",",r,")"), line=0.3,cex.main=1.2)


#posterior plot
plot(p, p.norm, type="l",lwd=2.5,col="darkgreen",xlim=c(0,1), ylim=c(0,1.1*max(p.norm)),
     xaxs="i",yaxs="i",xlab="p",ylab="P(p|r,n,M)")
title(main=paste("Posterior with (n,r)=(",n, ",",r,")"), line=0.3,cex.main=1.2)
```

# Exercise 2

Three  students  want  to  construct  their  prior  probability  about  the  proportion  of  residents  that support the building of a new concert hall in their small town.

- Anna thinks that her prior is a beta distribution with mean 0.2 and a standard deviation of 0.08.
- Benny moved only recently to this new town and therefore he does non have the slightest idea about it. Therefore he decides to use a uniform prior.
- Chris believes that his prior should have a trapezoidal shape:


 
 
 $$
f(X)=\left\{\begin{array}{ll}
20x \;\; 0x<0.1 \\ 2 \;\; 0.1x<0.3 \\ 510x \;\; 0.3x<0.5 \\ 0 \;\; x0.5
\end{array}\right.
$$




1. Draw and compare the three prior distributions.

The next day the three students decide to interview a sample of 100 citizens of the small town,asking for their opinion.  Out of the interviewed sample, 26 support the building of the new concert hall.

2. Evaluate and draw the three posterior distributions.

```r
n <- 100
r <- 26

options(repr.plot.width = 8, repr.plot.height = 7) 

#samples, delta and p 
n.sample <- 2000
delta.p <- 1/n.sample
p<-seq(from=1/(2*n.sample),by=1/n.sample,length.out=n.sample)


#ANNA 
m <- 0.2
sigma <- 0.08
alpha.prior <- m * (((m*(1-m))/sigma^2)-1)
beta.prior  <- (1-m)* (((m*(1-m))/sigma^2)-1)


prior.anna<-dbeta(x=p,alpha.prior,beta.prior)
likelihood.anna <-dbinom(x=r, size=n, prob=p)
p.norm.anna <-prior.anna*likelihood.anna/(delta.p*sum(likelihood.anna*prior.anna))
#posterior also a beta distib. dbeta(x=p, alpha.prior+r,beta.prior+n-r)


#BENNY
prior.benny <-dunif(p) 
likelihood.benny <-dbinom(x=r, size=n, prob=p) 
p.norm.benny <- prior.benny*likelihood.benny/(delta.p*sum(likelihood.benny*prior.benny))


#CHRIS
prior.chris.func <- function(x){
  prior <- ifelse( 0 <= x & x < 0.1, 20*x, ifelse(0.1 <= x & x < 0.3, 2, ifelse(0.3 <= x & x < 0.5, 5- 10*x , 0 )))
  return(prior)
    }

prior.chris <- prior.chris.func(p)
likelihood.chris <- dbinom(x=r, size=n, prob=p)
p.norm.chris <- prior.chris*likelihood.chris/(delta.p*sum(likelihood.chris*prior.chris))
z <-(delta.p*sum(likelihood.chris*prior.chris))

#PLOT PRIOR
par(mfrow=c(1, 2))
options(repr.plot.width = 15, repr.plot.height = 7) 
plot(p, prior.anna, type="l",xlim=c(0,1), ylim=c(0,6), ylab="P(p|M)",cex.axis=1, col="darkgreen", lwd=2)
title(main=paste("Comparision of the three prior distributions "), line=0.3,cex.main=1.2)
lines(p, prior.benny, type="l",lwd=2,col="navy",xlim=c(0,1), ylim=c(0,6))
lines(p, prior.chris.func(p), ylim=c(0,5),cex.axis=1, col="red", ylab="prior", lwd=2)
legend(x=0.5, y=5.5, legend=c("Anna's prior (Beta)", "Benny's prior (Uniform)", "Chris's prior (Trapezoidal)"),
       col=c("darkgreen", "navy", "red"), lty=1, cex=0.8)


#PLOT POSTERIOR
plot(p, p.norm.anna, type="l",xlim=c(0,1), ylim=c(0,11), ylab="P(p|n,r,M)",cex.axis=1, col="darkgreen", lwd=2)
title(main=paste("Comparision of the three posterior distributions "), line=0.3,cex.main=1.2)
lines(p, p.norm.benny, type="l",lwd=2, col="navy")
lines(p, p.norm.chris, type="l",lwd=2, col="red")
legend(x=0.6, y=10, legend=c("Anna's posterior", "Benny's posterior", "Chris's posteior"),
       col=c("darkgreen", "navy", "red"), lty=1, cex=0.8)

#FIX LEGEND
```

### 95% credibility interval

```r
cred_func <- function(posterior){
    p <- seq(from=1/(2*n.sample),by=1/n.sample,length.out=n.sample)
    p.mean <- p[which.max(posterior(p))]
    root <- uniroot(function(x) integrate(posterior,lower=p.mean-x, upper=p.mean+x)$value - 0.95, interval=c(0,1))$root
    low <- p.mean-root    
    upper <-p.mean+root  
    return(c(low,upper))
}

#Function vectorization for the uniroot and integral functions                    
#If we start with a Beta Prior with parameters alpha_p, beta_p and then measure r heads in n tosses
#the Posterior is a Beta functions with parameters alpha=alpha_p+r and beta=beta_p+n-r
post.func.anna <- function(p){dbeta(x=p, alpha.prior+r,beta.prior+n-r)}               
post.func.benny <-function(p){dbeta(x=p, 1+r,1+n-r)} #Prior Unif(0,1) = Beta(1,1) -> Posterior 1+r, 1+n-r 
#use of dcbinom package for the uniroot and integral functions  
post.func.chris <-function(p){post <- dcbinom(x=r, size=n, prob=p) * prior.chris.func(p)/z}

                   
cred.interval.anna <- cred_func(post.func.anna)
cred.interval.benny <- cred_func(post.func.benny)
cred.interval.chris <- cred_func(post.func.chris)

cat("The credibility interval are:")
cat("\n Anna: ",sprintf("%.2f", cred.interval.anna[1]),"-",sprintf("%.2f", cred.interval.anna[2]) )
cat("\n Benny: ",sprintf("%.2f", cred.interval.benny[1]),"-",sprintf("%.2f", cred.interval.benny[2]) )
cat("\n Chris: ",sprintf("%.2f", cred.interval.chris[1]),"-",sprintf("%.2f", cred.interval.chris[2]) )

# Plot the distribution and add the limits of the two CIs                   
options(repr.plot.width = 25, repr.plot.height = 7) 
par(mfrow=c(1, 3))
                    
    
plot(p, p.norm.anna, type="l",xlim=c(0,1), ylim=c(0,11), ylab="P(p|n,r,M)",cex.axis=1.5,cex.lab = 1.5, col="darkgreen", lwd=2)
title(main=paste("Credibility interval 95% - Anna posterior"), line=0.3,cex.main=2)                 
abline(v=cred.interval.anna[1],lty=2, lwd=1, col="darkgreen")
abline(v=cred.interval.anna[2],lty=2, lwd=1, col="darkgreen")
x.anna <-c(cred.interval.anna[1], seq(cred.interval.anna[1], cred.interval.anna[2],0.0001), cred.interval.anna[2])
y.anna <-c(0,post.func.anna( seq(cred.interval.anna[1], cred.interval.anna[2],0.0001)),0)
polygon(x.anna,y=y.anna, density = 20, angle = 45,col="darkgreen")     
text(x = cred.interval.anna[1]-0.02, y = 10, srt=90, col="darkgreen", label = paste("x =", sprintf("%.2f", cred.interval.anna[1])), cex = 2)
text(x = cred.interval.anna[2]+0.02, y = 10, col="darkgreen", label = paste("x =", sprintf("%.2f", cred.interval.anna[2])) ,
     srt = 0, cex = 2, xpd=TRUE, srt = 90)
                    
plot(p, p.norm.benny, type="l",lwd=2,col="navy",ylab="P(p|n,r,M)",cex.axis=1.5,cex.lab = 1.5)
title(main=paste("Credibility interval 95% - Benny posterior"), line=0.3,cex.main=2, cex.lab = 1.5, cex.axis=2)                 
abline(v=cred.interval.benny[1],lty=2, lwd=1, col="navy")
abline(v=cred.interval.benny[2],lty=2, lwd=1, col="navy") 
x.benny <-c(cred.interval.benny[1], seq(cred.interval.benny[1], cred.interval.benny[2],0.0001), cred.interval.benny[2])
y.benny <-c(0,post.func.benny( seq(cred.interval.benny[1], cred.interval.benny[2],0.0001)),0)
polygon(x.benny,y=y.benny, density = 20, angle = 45,col="navy")         
text(x = cred.interval.benny[1]-0.02, y = 8, srt=90, col="navy", label = paste("x =", sprintf("%.2f", cred.interval.benny[1])), cex = 2)
text(x = cred.interval.benny[2]+0.02, y = 8, col="navy", label = paste("x =", sprintf("%.2f", cred.interval.benny[2])) ,
     srt = 0, cex = 2, xpd=TRUE, srt = 90)


plot(p, post.func.chris(p), type="l",lwd=2, col="red", ylab="P(p|n,r,M)",cex.axis=1.5,cex.lab = 1.5)
title(main=paste("Credibility interval 95% - Chris posterior"), line=0.3,cex.main=2, cex.lab = 1.5, cex.axis=2)                 
abline(v=cred.interval.chris[1],lty=2, lwd=1, col="red")
abline(v=cred.interval.chris[2],lty=2, lwd=1, col="red")
x.chris <-c(cred.interval.chris[1], seq(cred.interval.chris[1], cred.interval.chris[2],0.0001), cred.interval.chris[2])
y.chris <-c(0,post.func.chris(seq(cred.interval.chris[1],cred.interval.chris[2],0.0001)),0)
polygon(x.chris,y=y.chris, density = 20, angle = 45,col="red")     
text(x = cred.interval.chris[1]-0.02, y = 8, srt=90, col="red", label = paste("x =", sprintf("%.2f", cred.interval.chris[1])), cex = 2)
text(x = cred.interval.chris[2]+0.02, y = 8, col="red", label = paste("x =", sprintf("%.2f", cred.interval.chris[2])) ,
     srt = 0, cex = 2, xpd=TRUE, srt = 90)

                    
#legend("topright", legend=c("Anna's posterior", "Benny's posterior", "Chris's posteior"),
#col=c("darkgreen", "navy", "red"), lty=1, cex=1)
```

2.  Evaluate the most probable value for the coin probability p and, integrating the posterior probability distribution, give an estimate for a 95% credibility interval.


```r
#Credibility interval 

cred_func <- function(posterior){
p <- seq(from=1/(2*n.sample),by=1/n.sample,length.out=n.sample)
p.mean <- p[which.max(posterior(p))]
root <- uniroot(function(x) integrate(posterior,lower=p.mean-x, upper=p.mean+x)$value - 0.95, interval=c(0,1))$root
low <-p.mean-root    
upper <-p.mean+root  
return(c(ifelse(low<=0,0,low),ifelse(upper<=0,0,upper),p.mean))
}
                
#Function vectorization for the uniroot and integral functions                    
#If we start with a Beta Prior with parameters alpha_p, beta_p and then measure r heads in n tosses
#the Posterior is a Beta functions with parameters alpha=alpha_p+r and beta=beta_p+n-r
post.func.beta <- function(p){dbeta(x=p, 10+r,10+n-r)}               
post.func.norm<-function(p){dbeta(x=p, 1+r,1+n-r)} #Prior Unif(0,1) = Beta(1,1) -> Posterior 1+r, 1+n-r 
                
cred.interval.norm <- cred_func(post.func.norm)
cred.interval.beta<- cred_func(post.func.beta)

#Mean value

cat("Estimate of the most probable value \n", "Flat:", sprintf("%.2f", cred.interval.norm[3]),
                                                "\n Beta:", sprintf("%.2f", cred.interval.beta[3]),"\n")
cat("Estimate for a 95% credibility interval \n", 
    "Flat:",  sprintf("%.2f",cred.interval.norm[1]), "-", sprintf("%.2f",cred.interval.norm[2]),
    "\n Beta:",  sprintf("%.2f",cred.interval.beta[1]), "-", sprintf("%.2f",cred.interval.beta[2]) )

options(repr.plot.width = 8, repr.plot.height = 7) 
                
plot(p, p.norm,type="l",lwd=2, ylim=c(0,8), ylab="P(r|p,n,M)", col="red", cex.axis=1,cex.lab = 1.5)
lines(p, p.norm.beta,type="l",lwd=2,col="navy")
title(main=paste("Posterior with (n,r) = (",n,",",r,")"), line=0.3,cex.main=1.5)

  
                
abline(v=cred.interval.norm[3],lty=2, lwd=1, col="red") 
abline(v=cred.interval.beta[3],lty=2, lwd=1, col="navy")
                
abline(v=cred.interval.norm[1],lty=2, lwd=1, col="red")
abline(v=cred.interval.norm[2],lty=2, lwd=1, col="red")  
                
abline(v=cred.interval.beta[1],lty=2, lwd=1, col="navy")
abline(v=cred.interval.beta[2],lty=2, lwd=1, col="navy") 
                
                
x.norm <-c(cred.interval.norm[1], seq(cred.interval.norm[1], cred.interval.norm[2],0.0001), cred.interval.norm[2])
y.norm <-c(0,post.func.norm( seq(cred.interval.norm[1], cred.interval.norm[2],0.0001)),0)
polygon(x.norm,y=y.norm, density = 15, angle = -45,col="red")  
text(x = cred.interval.norm[1]-0.02, y = 7, srt=90, col="red", label = paste("x =", sprintf("%.2f", cred.interval.norm[1])), cex = 1)
text(x = cred.interval.norm[2]+0.02, y = 7, col="red", label = paste("x =", sprintf("%.2f", cred.interval.norm[2])) ,
 srt = 0, cex = 1, xpd=TRUE, srt = 90)
                
                
                
x.beta <-c(cred.interval.beta[1], seq(cred.interval.beta[1], cred.interval.beta[2],0.0001), cred.interval.beta[2])
y.beta <-c(0,post.func.beta( seq(cred.interval.beta[1], cred.interval.beta[2],0.0001)),0)
polygon(x.beta,y=y.beta, density = 20, angle = 45,col="navy")         
text(x = cred.interval.beta[1]+0.02, y = 7, srt=90, col="navy", label = paste("x =", sprintf("%.2f", cred.interval.beta[1])), cex = 1)
text(x = cred.interval.beta[2]-0.02, y = 7, col="navy", label = paste("x =", sprintf("%.2f", cred.interval.beta[2])) ,
 srt = 0, cex = 1, xpd=TRUE, srt = 90)
                
```

3.  Repeat the same analysis assuming a sequential analysis of the data.  Show how the most probable value and the credibility interval change as a function of the number of coin tosses (i.e.  from 1 to30).

```r
outcomes <- c("T", "T", "T", "T", "T", "H","T", "T", "H", "H", "T", "T", "H","H", "H",
              "T", "H", "T", "H", "T", "H", "H", "T", "H", "T", "H", "T", "H", "H", "H")

n.sample <- 2000
delta.p <- 1/n.sample
p <- seq(from=1/(2*n.sample),by=1/n.sample,length.out=n.sample)


par(oma = c(4,1,1,1), mfrow = c(1, 5), mar = c(2, 2, 2, 2))
options(repr.plot.width = 25, repr.plot.height = 5) 


mean.flat <- c()
mean.beta <- c()
p.norm.plot <-c()
p.norm.beta.plot <- c()
cred.interv.norm.min <-c()
cred.interv.norm.max <-c()
cred.interv.norm.beta.min <-c()
cred.interv.norm.beta.max <-c()

cat("Plot of the posterior and credibility interval (95%) with a flat prior (red) and with a beta prior (navy) \n")


r=0
for(n in seq(from=1,to=30,by=1)){
    if(outcomes[n]=='H'){
        r = r+1
    }

    post.func.norm<-function(p){dbeta(x=p, 1+r,1+n-r)} #Prior Unif(0,1) = Beta(1,1) -> Posterior 1+r, 1+n-r 
    post.func.beta <- function(p){dbeta(x=p, 10+r,10+n-r)}               
    
    p.norm.plot <-append(p.norm.plot, p[which.max(post.func.norm(p))])
    p.norm.beta.plot <-append(p.norm.beta.plot, p[which.max(post.func.beta(p))])
      
    plot(p, post.func.norm(p),type="l",lwd=2, ylim=c(0,8), ylab="P(r|p,n,M)", col="red", cex.axis=2,cex.lab = 1.5)
    lines(p, post.func.beta(p),type="l",lwd=2,col="navy")
    title(main=paste("Posterior with (n,r) = (",n,",",r,")"), line=0.3,cex.main=2)
    
    mean.flat <-append(mean.flat, p[which.max(p.norm)])
    mean.beta <-append(mean.beta, p[which.max(p.norm.beta)])
    abline(v=p[which.max(post.func.norm(p))],lty=2, lwd=1, col="red") 
    abline(v=p[which.max(post.func.beta(p))],lty=2, lwd=1, col="navy") 

    cred.interval.norm <- cred_func(post.func.norm)
    cred.interval.beta<- cred_func(post.func.beta)

    x.beta <-c(cred.interval.beta[1], seq(cred.interval.beta[1], cred.interval.beta[2],0.0001), cred.interval.beta[2])
    y.beta <-c(0,post.func.beta( seq(cred.interval.beta[1], cred.interval.beta[2],0.0001)),0)
    polygon(x.beta,y=y.beta, density = 20, angle = 45,col="navy")         
                    
    x.norm <-c(cred.interval.norm[1], seq(cred.interval.norm[1], cred.interval.norm[2],0.0001), cred.interval.norm[2])
    y.norm <-c(0,post.func.norm(seq(cred.interval.norm[1], cred.interval.norm[2],0.0001)),0)
    polygon(x.norm,y=y.norm, density = 15, angle = -45,col="red")     
    
    cred.interv.norm.min <-append(cred.interv.norm.min,cred.interval.norm[1])
    cred.interv.norm.beta.min <-append(cred.interv.norm.beta.min,cred.interval.beta[1])
    cred.interv.norm.max <-append(cred.interv.norm.max, cred.interval.norm[2])
    cred.interv.norm.beta.max <-append(cred.interv.norm.beta.max,cred.interval.beta[2])

    }

add_legend <- function(...) {
  opar <- par(fig=c(0, 1, 0, 1), oma=c(0, 0, 0, 0), 
    mar=c(0, 0, 0, 0), new=TRUE)
  on.exit(par(opar))
  plot(0, 0, type='n', bty='n', xaxt='n', yaxt='n')
  legend(...)
}

add_legend("bottom", legend=c("Posterior with beta prior", "Posterior with flat prior"), 
   col=c("navy", "red"), lty=1,lwd=2,  horiz=TRUE, bty='n', cex=2)
```
