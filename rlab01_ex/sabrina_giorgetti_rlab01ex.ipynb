The following table gives the volume, area, length and maximum and mean depths of some Scottish lakes. Create  vectors,  holding  the  lake’s  name  and  all  the  parameters  and  build  a  dataframe called scottish.lakes from the vectors

```r
#Table gives the volume, area, length and maximum and mean depths of some Scottish lakes[1].
#Create vectors, holding the lake’s name and all the parameters and build a dataframe called scottish.lakes from the vectors

loch_names <- c("Loch Ness", "Loch Lomond", "Loch Morar", "Loch Tay", "Loch Awe", "Loch Maree", "Loch Ericht", "Loch Lochy", "Loch Rannoch","Loch Shiel", "Loch Katrine", "Loch Arkaig","Loch Shin") 
volume <- c(7.45, 2.6, 2.3, 1.6, 1.2, 1.09, 1.08, 1.07, 0.97, 0.79, 0.77, 0.75, 0.35)
area <- c(56, 71, 27, 26.4, 39, 28.6, 18.6, 16, 19, 19.5, 12.4, 16, 22.5)
length <- c(39 , 36 , 18.8 , 23 , 41 , 20 , 23 , 16 , 15.7 , 28 , 12.9 , 19.3 , 27.8)
max_depth <- c(230 , 190 , 310 , 150 , 94 , 114 , 156 , 162 , 134 , 128 , 151 , 109 , 49)
mean_depth <-c(132 , 37 , 87 , 60.6 , 32 , 38 , 57.6 , 70 , 51 , 40 , 43.4 , 46.5 , 15.5)
units <-c("-","km3", "km2", "km", "m", "m") #EVENTUALLY ADD UNITS

scottish.lakes <- data.frame(loch_names, volume, area, length, max_depth, mean_depth)
scottish.lakes
```

1. Evaluate the highest and lowest volume and area lake

```r
cat("The lake with the highest volume equal to", max(scottish.lakes$volume) ," km^3 is",loch_names[which.max(scottish.lakes$volume)] , "\n")
cat("The lake with the lowest volume equal to", min(scottish.lakes$volume) ,"km^3 is ", loch_names[which.min(scottish.lakes$volume)] , "\n")

cat("The lake with the highest area equal to", max(scottish.lakes$area) ,"km^2 is",loch_names[which.max(scottish.lakes$area)] , "\n")
cat("The lake with the lowest area equal to", min(scottish.lakes$area) ,"km^2 is", loch_names[which.min(scottish.lakes$area)] , "\n")
```

```r
ord_area_sl <- scottish.lakes[order(area, decreasing=TRUE),]
ord_area_sl[1:2,]

cat("The lakes with the two largest area are", ord_area_sl[1,1], "and",ord_area_sl[2,1])
cat("\n")
```

1. Write R code that is able to read the file and import it in a data frame structure.  (Hint:  before loading the file, open it with a text editor and check its structure)


```r
#The following CSV file https://drive.google.com/file/d/1cUQ8d0kDOw2lmzMAHIPglpdcnHy4GgF3/view 
#contains data on crude oil prices from 1861 to 2019, measured in US dollars per barrel.

id <- "1cUQ8d0kDOw2lmzMAHIPglpdcnHy4GgF3" # google file ID
crude_oil <- read.csv(sprintf("https://docs.google.com/uc?id=%s&export=download", id), skip = 1) #first row is gives a problem non having any columns, we cance skip it since doesn't contain useful info 
head(crude_oil)
```

2. Produce a plot with the Oil price as a function of the year


```r
options(repr.plot.width = 10, repr.plot.height = 6) 


pl_yearprice<- ggplot(crude_oil, aes(x=Year, y=OilPrice)) + geom_line(size=1) +
               labs(title="Oil price as a function of the year", x ="Year", y = "Oil Price")+
               theme(axis.text=element_text(size=14), plot.title=element_text(size=20,face="bold"), axis.title.x=element_text(size=16), axis.title.y=element_text(size=16)) 

pl_yearprice
```

4. Plot the derivative of the curve, simply evaluated with the finite difference formula (forward derivative)

```r
#∂price/year= pricej+1−pricej

dev <- vector(,nrow(crude_oil))
for (i in 0:nrow(crude_oil)){
    dev[i] <- crude_oil$OilPrice[i+1]- crude_oil$OilPrice[i]
}

options(repr.plot.width = 12, repr.plot.height = 6) 
dev_plt <- ggplot(crude_oil) + geom_line(aes(y = dev, x=1:length(dev)), color="blue") + labs(title="Oil Price derivative", x ="x", y = "y") +   geom_point(aes(y = dev, x=1:length(dev)),size=1) +
           theme(axis.text=element_text(size=14), plot.title=element_text(size=20,face="bold"), axis.title.x=element_text(size=16), axis.title.y=element_text(size=16)) 
dev_plt

```

## Exercise 3 - World Coal Production

1. Write R code that is able to read the file and import it in a tibble structure

```r
#The file contains data on the coal production (in TW·hour) for several countries in the World

id <- "10kgRQGKEC81QoVW9kj8goNzzyLcT0xld" # google file ID
coal <- read.csv(sprintf("https://docs.google.com/uc?id=%s&export=download", id))

coal_struct <- tibble(coal)
head(coal_struct)

```

2. Count the number of countries available in the file and produce a barplot with the number of entries for each country

```r
countries <- coal_struct %>% count(Entity) 
cat("The number of countries available in the file is equal to", length(countries$Entity))
cat("\n \n ")

options(repr.plot.width = 15, repr.plot.height = 35) 
plt <- ggplot(countries) + geom_col(aes(y = Entity, x=n)) + labs(title="Number of entries for each country", x ="Number of entries", y = "Country") +
       scale_x_continuous(expand = c(0,0.9), sec.axis = dup_axis()) +
       theme(axis.text=element_text(size=14), plot.title=element_text(size=20,face="bold"), axis.title.x=element_text(size=16), axis.title.y=element_text(size=16)) 
plt


```

3. Selecting only the year after 1970, determine the total integrated production for each country and print the top 5 Countries with highest coal productions

```r
coal_1970up <- filter(coal_struct, Year >= 1970)
coal_1970up_rank <- coal_1970up %>% group_by(Entity) %>% rename(production="Coal.production..TWh.") %>%
                summarise(production = sum(production)) %>% arrange(desc(production)) %>% slice(4,7,11,13,14) #to choose only the countries (not continent) 
coal_1970up_rank 
```

4. Selecting only the year after 1970, for the 5 top Countries, create a plot of production as a function of time

```r
options(repr.plot.width = 12, repr.plot.height = 10) 

coal_1970up_top5 <- filter(coal_1970up, Entity %in% coal_1970up_rank$Entity)
#coal_1970up_top5

plt_prod<- ggplot(coal_1970up_top5, aes(x=Year, y=Coal.production..TWh.)) + geom_point(mapping=aes(color=Entity), size=2) + scale_y_log10() + geom_line(mapping=aes(color=Entity))+
            labs(title="Plot of production as a function of time", x ="Year", y = "Coal.production TW·hour ")+
           theme(axis.text=element_text(size=14), plot.title=element_text(size=20,face="bold"), axis.title.x=element_text(size=16), axis.title.y=element_text(size=16)) 
plt_prod
```

5. Selecting only the year after 1970, generate a plot with the cumulative sum of the World’s coal production over the years

```r
coal_1970up_world <- coal_1970up %>% filter(Entity=="World") %>% rename(production="Coal.production..TWh.") %>% 
                     mutate(Cumulative_prod = cumsum(production) )

options(repr.plot.width = 10, repr.plot.height = 7) 

plt_prodW<- ggplot(coal_1970up_world, aes(x=Year, y=Cumulative_prod)) + 
            geom_line() + geom_point() + 
            labs(title="Cumulative sum of the World’s coal production over the years", x ="Year", y = "Cumulative production")+
            scale_fill_brewer(palette = "Spectral") +
           theme(axis.text=element_text(size=14), plot.title=element_text(size=20,face="bold"), axis.title.x=element_text(size=16), axis.title.y=element_text(size=16)) 
plt_prodW
```

#### Number of vaccinations in the various continents

```r
plot_func <- function(dataset,code, col){
             arg <- match.call()
             dataset <- filter(dataset, iso_code == arg$code)
             ggplot(dataset, aes(x=date, y = daily_vaccinations_per_million)) + geom_col(fill=arg$col) +
             facet_wrap(~location) + 
             labs(title="Number of daily vaccinations", x="Date (2020-2021)", y = "Number of vaccinations per million")+
             theme(axis.text=element_text(size=15), plot.title=element_text(size=20,face="bold"), 
                   axis.title.x=element_text(size=18), axis.title.y=element_text(size=18),
                   legend.title = element_text(size = 15), legend.text = element_text(size = 10),
                   strip.text = element_text(size = 25))  
    }



plt_vaccEUR <- plot_func(table, OWID_EUR, "#FF3300")
plt_vaccAFR <- plot_func(table, OWID_AFR, "#CC6633") 
plt_vaccNAM <- plot_func(table, OWID_NAM, "#0099FF")  
plt_vaccSAM <- plot_func(table, OWID_SAM, "#FF9900") 
plt_vaccASI <- plot_func(table, OWID_ASI, "#66CC33") 
plt_vaccOCE <- plot_func(table, OWID_OCE, "#FF6699") 


options(repr.plot.width = 35, repr.plot.height = 25) 
grid.arrange(plt_vaccEUR, plt_vaccAFR, plt_vaccSAM, plt_vaccNAM, plt_vaccASI, plt_vaccOCE,ncol=2)
```
